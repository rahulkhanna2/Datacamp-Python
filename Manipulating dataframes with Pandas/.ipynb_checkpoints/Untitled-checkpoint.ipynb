```python
import pandas as pd
import os
```

```python
election_file_path = os.path.join("pennsylvania2012_turnout.csv")
election_df = pd.read_csv(election_file_path, index_col ="county")

#Get only certain columns from election
results = election_df[["winner", "total", "voters"]]
(results.head(3))
```

```python
weather_file_path = os.path.join("pittsburgh2013.csv")
weather_df = pd.read_csv(weather_file_path)
weather_df.head(3)
```

```python
# Write a function to convert degrees Fahrenheit to degrees Celsius: to_celsius
def to_celsius(F):
    return 5/9*(F - 32)

# Apply the function over 'Mean TemperatureF' and 'Mean Dew PointF': df_celsius
df_celsius = weather_df[['Mean TemperatureF', 'Mean Dew PointF']].apply(to_celsius)

# Reassign the columns df_celsius
df_celsius.columns = ['Mean TemperatureC', 'Mean Dew PointC']

# Print the output of df_celsius.head()
(df_celsius.head())
```

```python
# Create the dictionary: red_vs_blue
red_vs_blue = {"Obama":"blue", "Romney":"red"}

# Use the dictionary to map the 'winner' column to the new column: election['color']
election_df['color'] = election_df.winner.map(red_vs_blue)

# Print the output of election.head()
print(election_df.head())
```

# Multi Indexing and Sorting

```python
sales = pd.DataFrame({"state":["CA","CA","NY","NY","TX","TX"], "month": [1, 2, 1, 2, 1, 2], 
                      "eggs":[ 47, 110, 221,  77, 132, 205]})
#Setting Multi index
sales = sales.set_index(["state", "month"])
sales
```

```python
sales = sales.sort_index()
(sales)
```

```python
# Look up data for NY in month 1: NY_month1
NY_month1 = sales.loc[('NY', 1), :]
print(NY_month1)

# Look up data for CA and TX in month 2: CA_TX_month2
CA_TX_month2 = sales.loc[(['CA', 'TX'], 2), :]
print (CA_TX_month2)

# Look up data for all states in month 2: all_month2
all_month2 = sales.loc[(slice(None), 2), :]
print (all_month2)
```
